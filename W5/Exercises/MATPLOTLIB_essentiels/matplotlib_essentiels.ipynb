# 1. üìä Pourquoi visualiser des donn√©es ?

### Cas concret
Vous pouvez explorer la courbe de variation d'une variable num√©rique dans un jeu de donn√©es (ex : temp√©rature journali√®re, revenu mensuel, stock d‚Äôun produit).

Ci-dessous, un exemple d'une visualisation de l'√©volution d'un compte d'√©pargne fictif.

```python
import matplotlib.pyplot as plt

# Cr√©ation d'une liste d'ann√©es allant de 1 √† 10
annees = list(range(1, 11)) 

# Calcul du solde du compte √©pargne pour chaque ann√©e, avec un taux d‚Äôint√©r√™t de 5 % par an
# Formule utilis√©e : capital initial (1000 ‚Ç¨) * (1.05)^nombre d'ann√©es
solde = [1000 * (1.05)**i for i in annees]      
solde_2 = [1000 * (1.03)**i for i in annees]      

# Trac√© de la courbe repr√©sentant l'√©volution de l'√©pargne,
# avec une √©tiquette ("label") qui sera utilis√©e dans la l√©gende du graphique
# ‚ÑπÔ∏è il est possible d'ajouter plusieurs courbes sur le m√™me sch√©ma en appelant la m√©thode .plot() pour chaque  courbe.
plt.plot(annees, solde, label="√âpargne (5%/an)")

# Ajout du label de l‚Äôaxe des x (les ann√©es)
plt.xlabel("Ann√©es")

# Ajout du label de l‚Äôaxe des y (le solde en euros)
plt.ylabel("Solde (‚Ç¨)")

# Titre du graphique
plt.title("√âvolution de l'√©pargne sur 10 ans") 

# Affichage de la l√©gende pour identifier la courbe
plt.legend()

# Affichage d‚Äôune grille pour faciliter la lecture du graphique
plt.grid(True)

# Affichage du graphique final
plt.show()
```

# 2. üìà Utiliser le bon type de graphique

En **exploration de donn√©es**, le type de graphique utilis√© d√©pend de la nature des donn√©es

### 2.1. Pour les **donn√©es quantitatives**

#### a. Courbe (line plot)

Un graphique en courbe est particuli√®rement adapt√© pour suivre l‚Äô√©volution d‚Äôune variable dans le temps, car il permet de visualiser les tendances, les hausses et les baisses de mani√®re continue.

*(voir les graphiques pr√©c√©dents)*

#### b. Histogramme

Un histogramme est utilis√© pour observer la distribution d‚Äôune variable continue, comme la taille ou le revenu, afin de visualiser comment les valeurs se r√©partissent sur diff√©rents intervalles.

```python
# Liste des notes obtenues par des √©l√®ves
notes = [12, 15, 14, 17, 10, 11, 18, 13, 14, 15, 16, 8, 12, 13, 16, 7, 8, 10, 12]

# Cr√©ation de l'histogramme :
# - bins=5 : nombre d'intervalles pour regrouper les notes
# - rwidth=0.8 : largeur relative des barres (pour laisser de l'espace entre elles)
plt.hist(notes, bins=5, rwidth=0.8)

# Ajout du titre du graphique
plt.title("Distribution des notes")

# √âtiquette de l'axe des x : les classes de notes
plt.xlabel("Notes")

# √âtiquette de l'axe des y : nombre d'√©l√®ves par intervalle de notes
plt.ylabel("Fr√©quence")

# Affichage du graphique
plt.show()
```

### 2.2. Pour les **donn√©es cat√©gorielles**

#### a. Diagramme en barres (bar plot)

On utilise un diagramme en barres pour comparer le nombre d‚Äôobservations entre diff√©rentes cat√©gories, par exemple le nombre de clients par r√©gion ou les ventes enregistr√©es chaque mois.

```python
# Donn√©es : noms des mois
mois = ["Jan", "F√©v", "Mar", "Avr", "Mai"]

# Donn√©es : nombre de ventes correspondantes √† chaque mois
ventes = [100, 150, 90, 200, 130]

# Cr√©ation d‚Äôun diagramme en barres pour visualiser les ventes par mois
plt.bar(mois, ventes)

# Ajout d‚Äôun titre au graphique
plt.title("Ventes mensuelles")

# Affichage du graphique
plt.show()
```

**Camembert (pie chart)** : pour visualiser la r√©partition d‚Äôune variable qualitative (ex : r√©partition des ventes par produit).

```python
# D√©finition des parts de march√© en pourcentage
parts = [40, 30, 20, 10]

# √âtiquettes correspondant √† chaque part (nom des entreprises)
labels = ['Entreprise A', 'B', 'C', 'D']

# Cr√©ation du graphique en camembert (pie chart)
# - labels : pour afficher les noms des entreprises
# - autopct : pour afficher les pourcentages avec une d√©cimale
plt.pie(parts, labels=labels, autopct='%1.1f%%')

# Ajout d‚Äôun titre au graphique
plt.title("Parts de march√©")

# Affichage du graphique
plt.show()
```

### 2.3. Pour √©tudier la **relation entre deux variables**

**Nuage de points (scatter plot)** : pour analyser la relation entre deux variables, par exemple avant de r√©aliser une r√©gression.

```python
# D√©finition des coordonn√©es en abscisses (x)
x = [1, 2, 3, 4, 5]

# D√©finition des coordonn√©es en ordonn√©es (y) associ√©es √† chaque point x
y = [2.2, 2.8, 4.5, 3.7, 5.0]

# Cr√©ation d'un nuage de points (scatter plot) pour visualiser la relation entre x et y
plt.scatter(x, y)

# Titre du graphique
plt.title("Corr√©lation X vs Y")

# Affichage du graphique
plt.show()
```

# 3. üìàüìâ Comparer plusieurs graphiques dans une m√™me figure

Lorsque vous analysez un dataset complexe, vous devez souvent explorer plusieurs dimensions en parall√®le. 
Pouvoir afficher plusieurs visualisations sur une seule **figure** facilite la comparaison visuelle et cela gr√¢ce √† la m√©thode `subplot`.

```python
# Cr√©e une figure avec une taille personnalis√©e (largeur = 10, hauteur = 8 pouces)
plt.figure(figsize=(10,8))

# Premier graphique (en haut √† gauche) : diagramme en barres
plt.subplot(2,2,1)                                  # 2 lignes, 2 colonnes, 1er sous-graphe
plt.bar(mois, ventes)                               # Affiche les ventes par mois sous forme de barres
plt.title("Barres")                                 # Titre du graphique

# Deuxi√®me graphique (en haut √† droite) : nuage de points (scatter plot)
plt.subplot(2,2,2)                                  # 2 lignes, 2 colonnes, 2e sous-graphe
plt.scatter(x, y)                                   # Affiche la relation entre deux variables
plt.title("Scatter")                                # Titre du graphique

# Troisi√®me graphique (en bas √† gauche) : histogramme
plt.subplot(2,2,3)                                  # 2 lignes, 2 colonnes, 3e sous-graphe
plt.hist(notes, bins=5, rwidth=0.8)                 # Montre la distribution des notes
plt.title("Histogramme")                            # Titre du graphique

# Quatri√®me graphique (en bas √† droite) : diagramme circulaire (camembert)
plt.subplot(2,2,4)                                  # 2 lignes, 2 colonnes, 4e sous-graphe
plt.pie(parts, labels=labels, autopct='%1.1f%%')    # R√©partition des parts avec pourcentages
plt.title("Camembert")                              # Titre du graphique

# Ajuste automatiquement les espacements entre les sous-graphes pour √©viter les chevauchements
plt.tight_layout()

# Affiche l‚Äôensemble des graphiques
plt.show()
```

# 4. üß© Travailler avec un dataset r√©el

Le dataset contient des mesures de fleurs d‚Äôiris appartenant √† trois esp√®ces diff√©rentes :
- Longueur et largeur des **s√©pales**
- Longueur et largeur des **p√©tales**
- Esp√®ce de la fleur (**setosa**, **versicolor**, **virginica**)

>
> Affichez un histogramme de la longueur des s√©pales pour visualiser la distribution de cette variable continue.

> Utilisez un diagramme en barres pour afficher la r√©partition des esp√®ces.
>
> Utilisez un nuage de points pour √©tudier la relation entre la longueur du s√©pale et celle du p√©tale

```python
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.hist(df["sepal_length"], bins=5, rwidth=0.8)
plt.title("Distribution de la longueur des s√©pales")
plt.xlabel("Longueur des s√©pales")
plt.ylabel("Fr√©quence")

species_counts = df["species"].value_counts()

plt.subplot(2,2,2)
plt.bar(species_counts.index, species_counts.values)
plt.title("Distribution des esp√®ces")
plt.xlabel("Esp√®ce")
plt.ylabel("Fr√©quence")

plt.subplot(2,2,3)
plt.scatter(df["sepal_length"], df["petal_length"])
plt.title("Corr√©lation longueur des s√©pales/longueur des p√©tales")
plt.xlabel("Longueur des s√©pales")
plt.ylabel("Longueur des p√©tales")

plt.tight_layout()

plt.show()

df.head()
```
